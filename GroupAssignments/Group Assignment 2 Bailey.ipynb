## Mapping the final data analysis with basemap
To start off, let's visualize the MHI of every census tract in the whole city of LA. We'll use custom (user-defined) brackets to highlight in purple the census tracts with a MHI lower than \\$25750.

```python
mhi.plot(figsize=(15,15), 
         column='Median Household Income', 
         cmap = 'Spectral_r', 
         legend = True, 
         scheme = 'user_defined', 
        classification_kwds={'bins':[25750,40000,60000,800000,100000,150000,200000,250000]}, 
        edgecolor = 'white', 
        linewidth = 0.1, 
        )
```

Next, we'll add a basemap below the plot to give the plot a geospatial context of its surrounding environment.

```python
mhi_web_mercator = mhi.to_crs(epsg=3857)
```

```python
fig, ax = plt.subplots(figsize= (15,15))

mhi_web_mercator[mhi_web_mercator['Median Household Income'] > 0].plot(
    ax = ax,
    column= 'Median Household Income', 
    cmap = 'Spectral_r', 
    legend = True, 
    scheme = 'user_defined', 
    classification_kwds={'bins':[25750,40000,60000,800000,100000,150000,200000,250000]}, 
    edgecolor = 'white', 
    linewidth = 0.1, 
    alpha= 0.8
)

ax.axis('on')

ax.set_title('City of Los Angeles Census Tracts by \n Median Household Income (USD)', fontsize=24, pad=20)

ctx.add_basemap(ax)
```

Awesome! Now we'll create two side-by-side graphs of LA: one showing the census tracts with MHI lower than the federal poverty guideline for a family of four, and one showing the top twenty census tracts by MHI. I use a black-and-white basemap to highlight the locations of the census tracts. Now I can confirm my previous assumptions about where these census tracts were located are correct.

```python
#creating the side-by-side subplots (1x2)
fig, axs = plt.subplots(1, 2, figsize=(15,15), sharex=True, sharey=True)

#naming each subplot
ax1, ax2 = axs

#adding the layer with ax=ax in the argument for the first map
mhi_web_mercator[mhi_web_mercator['Median Household Income'] < 25750].plot(ax=ax1,
                                                                           alpha = 0.5,
                                                                           color = "purple",
                                                                          edgecolor = "white",
                                                                          linewidth = 0.2)

#setting manual x and y limits to match the City of LA zoom level rather than zooming down to just the census tracts
ax1.set_title('LA Census Tracts with Median Household Income Below \n the Federal Poverty Guideline for a Family of Four', fontsize=15, pad=20)
ax1.set_ylim([3980000,4088000])
ax1.set_xlim([-13207500,-13140000])

#adding first basemap
ctx.add_basemap(ax1,source=ctx.providers.Stamen.Toner)

#adding the layer with ax=ax in the argument for the second map
mhi_web_mercator[mhi_web_mercator['Median Household Income']>188408.9].plot(ax=ax2,
                                                                                         alpha = 0.5,
                                                                                         color = "red",
                                                                                      edgecolor = "white",
                                                                                      linewidth = 0.2)

ax2.set_title('Top Twenty LA Census Tracts by Median Household Income', fontsize=15, pad=20)

#adding second basemap
ctx.add_basemap(ax2, source=ctx.providers.Stamen.Toner)
```

Since zooming out to the entire City of LA makes it hard to see the individual census tracts, I'll make another side-by-side map that turns off the manual x and y limits as well as the shared x and y axes so that the plot automatically zooms in to see both maps at a closer level. 

```python
#creating the side-by-side subplots (1x2)
fig, axs = plt.subplots(1, 2, figsize=(15,15), sharex=False, sharey=False)

#naming each subplot
ax1, ax2 = axs

#adding the layer with ax=ax in the argument for the first map
mhi_web_mercator[mhi_web_mercator['Median Household Income'] < 25750].plot(ax=ax1,
                                                                           alpha = 0.5,
                                                                           color = "purple",
                                                                          edgecolor = "white",
                                                                          linewidth = 0.2)

#setting x and y limits automatically by removing 'set_ylim' and 'set_xlim' commands
ax1.set_title('LA Census Tracts with Median Household Income Below \n the Federal Poverty Guideline for a Family of Four', fontsize=15, pad=20)
#ax1.set_ylim([3980000,4088000])
#ax1.set_xlim([-13207500,-13140000])

#adding first basemap
ctx.add_basemap(ax1,source=ctx.providers.Stamen.Toner)

#adding the layer with ax=ax in the argument for the second map
mhi_web_mercator[mhi_web_mercator['Median Household Income']>188408.9].plot(ax=ax2,
                                                                                         alpha = 0.5,
                                                                                         color = "red",
                                                                                      edgecolor = "white",
                                                                                      linewidth = 0.2)

ax2.set_title('Top Twenty LA Census Tracts by Median Household Income', fontsize=15, pad=20)

#adding second basemap
ctx.add_basemap(ax2, source=ctx.providers.Stamen.Toner)
```

Notes of interest: The plots surrounding LA's two largest universities--UCLA and USC--contain some of the lowest MHI census tracts in the city. This might be due to the number of students without income reporting as independent households in these census tracts.
